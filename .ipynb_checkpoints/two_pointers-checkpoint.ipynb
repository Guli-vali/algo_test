```python
"""

Основные типы:
    
Встречные указатели (left ← → right)
Один слева, другой справа, движутся навстречу
Примеры: палиндромы, поиск пары с суммой

Читающий и пишущий (read → write)
Один читает все, другой пишет только нужное
Примеры: перемещение нулей, удаление дубликатов

Быстрый и медленный (slow → fast)
Один движется быстрее (например, в 2 раза)
Примеры: поиск середины, циклов

Два указателя в разных структурах (i → j)
Каждый в своем массиве/списке
Примеры: объединение отсортированных массивов

Преимущества:
Эффективность: часто O(n) вместо O(n²)
Экономия памяти: O(1) дополнительной памяти
Простота реализации

"""
```

```python
"""
ID: 34
Тема: Two pointers
Название: Переместить нули в конец
Сложность: Easy
--------------------------------------------------------------------------------
Условие:
Дан массив. Переместить все нули в конец, сохранив относительный порядок остальных элементов.

Пример:
Вход: [0, 1, 0, 3, 12]
Выход: [1, 3, 12, 0, 0]
"""
```

```python
def zeros_switch(arr):
    write_ix = 0
    for read_ix in range(len(arr)):
        if arr[read_ix] != 0:
            arr[write_ix] = arr[read_ix]
            write_ix += 1

    while write_ix < len(arr):
        arr[write_ix] = 0
        write_ix += 1

    return arr

print(zeros_switch([0, 1, 0, 3, 12]))
```

```python
"""
ID: 33
Тема: Two pointers
Название: Удалить дубликаты из отсортированного массива
Сложность: Easy
--------------------------------------------------------------------------------
Условие:
Дан отсортированный массив. Удалить дубликаты на месте, вернув новую длину.

Пример:
Вход: [1, 1, 2, 2, 3, 4, 4, 5]
Выход: 5 (массив становится [1, 2, 3, 4, 5, ...])
"""
```

```python
def remove_dubs(arr):
    write_ix = 0
    read_ix = 0

    while read_ix < len(arr):
        if arr[read_ix] != arr[write_ix]:
            read_ix += 1
        arr.pop(read_ix)
        write_ix = +1

    return arr
        
            

print(remove_dubs([1, 1, 2, 2, 3, 4, 4, 5]))
```
